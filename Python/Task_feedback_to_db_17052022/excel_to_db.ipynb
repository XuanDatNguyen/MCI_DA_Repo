```python
# Create the INSERT INTO sql query
query = """INSERT INTO hocvien_studentfeedback
                        (created, name, domain, unsatisfaction_thing, building_comment, understanding_comment,
                        approachable_level, continued, expertise_level, enthusiastic_level, understanding_level, 
                        satisfaction_level, class_code, lecturer_name, feedback_time, lecturer_id)
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s,
                        %s, %s, %s, %s, %s, %s, %s, %s)"""

# Create a For loop to iterate through each row in the XLS file, starting at row 2 to skip the headers
count_all_single_course = 0

for r in range(feedback_df.shape[0]):
    created               = feedback_df.created[r]
    name                  = feedback_df.name[r]
    domain                = feedback_df.domain[r]
    unsatisfaction_thing  = feedback_df.unsatisfaction_thing[r]
    building_comment      = feedback_df.building_comment[r]
    understanding_comment = feedback_df.understanding_comment[r]
    approachable_level    = feedback_df.approachable_level[r]
    continued             = feedback_df.continued[r]
    expertise_level       = feedback_df.expertise_level[r]
    enthusiastic_level    = feedback_df.enthusiastic_level[r]
    understanding_level   = feedback_df.understanding_level[r]
    satisfaction_level    = feedback_df.satisfaction_level[r]
    class_code            = feedback_df.class_code[r]
    lecturer_name         = feedback_df.lecturer_name[r]
    feedback_time         = feedback_df.feedback_time[r]
    lecturer_id           = feedback_df.lecture_id[r]
    
    # Assign values from each row
    values = (created, name, domain, unsatisfaction_thing, building_comment, understanding_comment, 
              approachable_level, continued, expertise_level, enthusiastic_level, understanding_level, 
              satisfaction_level, class_code, lecturer_name, feedback_time, lecturer_id)

    # Execute sql Query
    cursor.execute(query, values)
    count_all_single_course += 1

# Close the cursor
#cursor.close()

# Commit the transaction
database.commit()

# Close the database connection
database.close()

# Print results
print ("")
print ("All Done! Bye, for now.")
print ("")
print ("I just imported " + str(count_all_single_course) + " rows to MySQL!")
```
